```julia
begin
    tree = ((:a, :b), (:c, (:d, :e)))

    fig = Figure()
    ax = Axis(fig[1, 1], )
    hidedecorations!(ax)
    hidespines!(ax)
    treearea!(tree; addstroke=true, strokestyle=:dashdot, strokewidth=1.0, strokecolor=:magenta, padding=((0.0, 0.0), (0.0, 0.0)))
    treearea!(tree; node=(:c, (:d, :e)), color=:green, alpha=0.2)
    treearea!(tree; node=(:d, :e), color=:green, alpha=0.2)
    treearea!(tree; node=(:a, :b), color=:red, alpha=0.2)
    treelabels!(tree, Dict(:a => "A", :b => "B", :c => "C", :d => "D", :e => "E"), usemaxdepth=true)
    treescatter!(tree)
    treecladelabel!(tree, node=(:d, :e), color=:green, font=:bold, label="Even-Cooler-Clade", lineoffset=0.4,    rotation=pi/2, align=(:center, :top))
    treecladelabel!(tree, node=(:c, (:d, :e)), color=:lime, label="Clade", lineoffset=0.2, linestyle=:dash, linewidth=3,   rotation=pi/2, align=(:center, :top))
    treeplot!(tree)
    # rotate!(ax.scene, deg2rad(-90))
    fig
end
```

```julia
begin
    tree = ((:a, :b), (:c, (:d, :e)))

    fig = Figure(size=(600,600))
    ax = PolarAxis(fig[1, 1], 
        # xautolimitmargin=(0.1, 0.5)
        clip=false,
    )
    hidedecorations!(ax)
    hidespines!(ax)
    tp = treeplot!(tree)
    treearea!(tree; 
        addstroke=true, strokestyle=:dashdot, strokewidth=1.0, strokecolor=:magenta, 
        padding=((0.0, 0.0), (0.0, 0.0)), resolution=50
    )
    treearea!(tree; node=(:c, (:d, :e)), color=:green, alpha=0.2,)
    treearea!(tree; node=(:d, :e), color=:green, alpha=0.2, )
    treearea!(tree; node=(:a, :b), color=:red, alpha=0.2)
    treelabels!(tree, Dict(:a => "A", :b => "B", :c => "C", :d => "D", :e => "E"), 
        usemaxdepth=true, 
        tiplabeloffset=(1.0, 0.0)
    )
    treescatter!(tree)
    treecladelabel!(tree, node=(:d, :e), color=:green, font=:bold, label="Super-Interesting-Clade", 
        lineoffset=1.0, linestyle=:dash, linewidth=1, rotation=pi/2, align=(:center, :top),
    )
    treecladelabel!(tree, node=(:c, (:d, :e)), color=:lime, label="Clade", 
        lineoffset=0.5, linestyle=:dash, linewidth=3, rotation=pi/2, align=(:center, :top),
    )
    # p = treeplot!(tree)
    fig
end
```
